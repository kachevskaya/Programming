```python
# №1

import threading
import time

def numbers():
    for i in range (1,11):
        print (i)
        time.sleep(0.1)

def letters():
    for i in range(65, 75):
        print(chr(i))
        time.sleep(0.1)
    
thread_1 = threading.Thread(target=numbers)
thread_2 = threading.Thread(target=letters)

thread_1.start()
thread_2.start()

thread_1.join()
thread_2.join()
```

```python
# №2

import multiprocessing

def factorial(n):
    if n ==1:
        return 1
    else:
        return n*factorial(n-1)

numbers = [1,2]
with multiprocessing.Pool() as pool:
    results = pool.map(factorial, numbers)
print(results)
```

```python
# №3

import threading

def count_words(text_name, text, result_dict):
    words = text.split()
    result_dict[text_name] = len(words)

if __name__ == "__main__":
    texts = {"text1": "Пример текста", "text2": "Еще один текст"}
    
    results = {}
    threads = []
    
    for text_name, text in texts.items():
        thread = threading.Thread(target=count_words, args=(text_name, text, results))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
    
    for text_name, word_count in results.items():
        print(f"Количество слов в {text_name}: {word_count}")
```

```python
# №4

import multiprocessing

def calculate_fibonacci(n):
    def fib(n):
        if n <= 1:
            return n
        else:
            return fib(n - 1) + fib(n - 2)
    return n, fib(n)

if __name__ == "__main__":
    
    numbers = [10, 20, 30, 40]
    
    with multiprocessing.Pool() as pool:
        results = pool.map(calculate_fibonacci, numbers)

    for number, fibonacci in results:
        print(f"Число Фибоначчи для {number} равно {fibonacci}")
```

```python
import threading
import multiprocessing
import time

def calculate_square_sum(number):
    return sum(i * i for i in range(number))

def threading_approach(numbers):
    results = {}
    threads = []

    def worker(number):
        results[number] = calculate_square_sum(number)

    for number in numbers:
        thread = threading.Thread(target=worker, args=(number,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    return results


def multiprocessing_approach(numbers):
    with multiprocessing.Pool() as pool:
        results = pool.map(calculate_square_sum, numbers)
    return dict(zip(numbers, results))

if __name__ == "__main__":
    numbers = [500, 600, 700, 800]

   
    start_time = time.time()
    threading_results = threading_approach(numbers)
    threading_time = time.time() - start_time

    
    start_time = time.time()
    multiprocessing_results = multiprocessing_approach(numbers)
    multiprocessing_time = time.time() - start_time

   
    print("threading:")
    for number, result in threading_results.items():
        print(f"Сумма квадратов для {number}: {result}")
    print(f"Время выполнения (threading): {threading_time:.4f} секунд")

    print("multiprocessing:")
    for number, result in multiprocessing_results.items():
        print(f"Сумма квадратов для {number}: {result}")
    print(f"Время выполнения (multiprocessing): {multiprocessing_time:.4f} секунд")
```
