```python
# №1

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C']
}

def dfs(graph, start, visited=None):
    
    if visited is None:
        visited = []
        
    visited.append(start)
    
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited

start_node = 'A'
dfs_order = dfs(graph, start_node)

print( dfs_order)
```

```python
# №2
# Класс для узла дерева
class Node:
    def __init__(self, value):
        self.value = value  
        self.left = None    
        self.right = None   

# Класс для самого дерева
class BinarySearchTree:
    def __init__(self):
        self.root = None  

    # Метод вставки нового значения в дерево
    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self.insert_recursive(self.root, value)

    # Вспомогательный метод для рекурсивной вставки
    def insert_recursive(self, current_node, value):
        if value < current_node.value:
            if current_node.left is None:
                current_node.left = Node(value)
            else:
                self.insert_recursive(current_node.left, value)
        else:
            if current_node.right is None:
                current_node.right = Node(value)
            else:
                self.insert_recursive(current_node.right, value)

    # Метод инфиксного (inorder) обхода
    def inorder_traversal(self):
        result = []
        self.inorder_recursive(self.root, result)
        return result

    # Вспомогательная рекурсивная функция для обхода
    def inorder_recursive(self, node, result):
        if node:
            
            self.inorder_recursive(node.left, result)
            result.append(node.value)
            self.inorder_recursive(node.right, result)


bst = BinarySearchTree()
elements = [10, 5, 15, 3, 7, 12, 18]

for elem in elements:
    bst.insert(elem)

inorder_result = bst.inorder_traversal()

print("Инфиксный обход:", inorder_result)
```

```python
# №3
from collections import deque


queue = deque()
stack = deque()

elements = [1, 2, 3, 4]

for elem in elements:
    queue.append(elem)     
    stack.append(elem)     

queue_output = []
while queue:
    queue_output.append(queue.popleft())  

stack_output = []
while stack:
    stack_output.append(stack.pop())     

print("Порядок возвращения элементов очереди:", queue_output)
print("Порядок возвращения элементов стэка:", stack_output)
```

```python
# №4


class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key) 
        
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value) 
                return "updated"
            
        self.table[index].append((key, value))
        return "added"

    def get(self, key):

        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None 

    def delete(self, key):

        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i] 
                return True
        return False  

# Функция для текстового меню
def menu():
    ht = HashTable()  # Создаём хэш-таблицу
    while True:
        print("\nВыберите действие:")
        print("1. Вставить ключ-значение")
        print("2. Получить значение по ключу")
        print("3. Удалить ключ")
        print("4. Выйти")
        choice = input("Ваш выбор: ")

        if choice == "1":
            key = input("Введите ключ: ")
            value = input("Введите значение: ")
            result = ht.insert(key, value)
            if result == "added":
                print(f" Пара ('{key}': '{value}') добавлена.")
            elif result == "updated":
                print(f" Значение ключа '{key}' обновлено на '{value}'.")

        elif choice == "2":
            key = input("Введите ключ: ")
            value = ht.get(key)
            if value is not None:
                print(f" Значение по ключу '{key}': {value}")
            else:
                print(f" Ключ '{key}' не найден.")

        elif choice == "3":
            key = input("Введите ключ: ")
            if ht.delete(key):
                print(f" Ключ '{key}' удалён.")
            else:
                print(f" Ключ '{key}' не найден.")

        elif choice == "4":
            print(" Выход из программы.")
            break

        else:
            print(" Неверный выбор. Попробуйте снова.")

# Запуск меню:
menu()
```

```python
# №5

class Node:
    def __init__(self, value):
        self.value = value    
        self.next = None      

class LinkedList:
    def __init__(self):
        self.head = None  

    def append(self, value):
        new_node = Node(value)  
        if self.head is None:
            self.head = new_node  
        else:
            current = self.head
            while current.next: 
                current = current.next
            current.next = new_node 

    def display(self):
        elements = []
        current = self.head
        while current: 
            elements.append(str(current.value))
            current = current.next
        return " -> ".join(elements)


linked_list = LinkedList()
for value in [1, 2, 3, 4]:
    linked_list.append(value)

print("Односвязный список:", linked_list.display())
```
